<div style="font-family: 'B Nazanin', serif; direction: rtl; border: 1px solid #ccc; border-radius: 5px; padding: 10px; font-size: 18px;">
    <h2>آزمون t مستقل :</h2>
    <p>در اینجا یک آزمون تی-مستقل بین دو گروه مختلف از داده‌ها انجام می‌شود. فرض صفر در اینجا این است که میانگین دو گروه برابر است. به طور خاص، از تابع <code>ttest_ind</code> از کتابخانه <code>scipy.stats</code> استفاده شده است.</p>
    <ul>
        <li><strong>test:</strong> نتیجه آزمون t را نشان می‌دهد.</li>
        <li><strong>pval:</strong> مقدار P-Value مربوط به آزمون t را نشان می‌دهد.</li>
    </ul>
   

<h3>نتیجه آزمون t در test ذخیره می‌شود و مقدار P-Value مربوطه در pval ذخیره می‌شود. این دو مقدار به طور معمول برای ارزیابی تفاوت معنادار بین میانگین دو گروه استفاده می‌شوند. اگر مقدار P-Value کمتر از سطح معناداری مشخص شده باشد (مثلا 0.05)، ما می‌توانیم فرض صفر را رد کنیم و به این نتیجه برسیم که تفاوت معناداری بین میانگین دو گروه وجود دارد.</h3>
</div>

$$H_{0}  :  μ_{A} =  μ_{B}$$
$$H_{1}  :  μ_{A} \neq  μ_{B}$$

```python
test,pval = ttest_ind(boys['stress_first'],girls['stress_first'])
print('test :' , test)
print('P-value :' , pval)
```

```python
test,pval = ttest_ind(boys['num_ill'],girls['num_ill'])
print('test :' , test)
print('P-value :' , pval)
```

```python
test,pval = ttest_ind(boys['grade_pre'],girls['grade_pre'])
print('test :' , test)
print('P-value :' , pval)
```

```python
test,pval = ttest_ind(boys['final_grade'],girls['final_grade'])
print('test :' , test)
print('P-value :' , pval)
```

```python
test,pval = ttest_ind(boys['stress_final'],girls['stress_final'])
print('test :' , test)
print('P-value :' , pval)
```

<div style="font-family: 'B Nazanin', serif; direction: rtl; border: 1px solid #ccc; border-radius: 5px; padding: 10px; font-size: 18px; text-align: right;">
    <h2>آزمون t زوجی :</h2>
    <p>در اینجا به دلیل اینکه همان دانشجویان  استرسشان در ابتدای سال و انتهای سال اندازه گرفته شده است مستقل نیستند و باید از آزمون t زوجی استفاده کنیم </p>
    <ul>
        <li><code>df['stress_first']</code>: نشان دهنده داده‌های گروه یا شرایط قبلی است.</li>
        <li><code>df['stress_final']</code>: نشان دهنده داده‌های گروه یا شرایط بعدی است.</li>
    </ul>
    <p>و نتیجه این آزمون به صورت دو مقدار <code>test</code> و <code>pval</code> بازگردانده می‌شود:</p>
    <ul>
        <li><strong>test:</strong> نتیجه آزمون t را نشان می‌دهد.</li>
        <li><strong>pval:</strong> مقدار P-Value مربوط به آزمون t را نشان می‌دهد.</li>
    </ul>
    <p>مقدار P-Value معمولاً برای ارزیابی معناداری تفاوت بین میانگین دو گروه یا شرایط متناظر استفاده می‌شود. اگر P-Value کمتر از سطح معناداری مشخص شده باشد (مثلاً 0.05)، می‌توانیم فرض صفر را رد کنیم و به این نتیجه برسیم که تفاوت معناداری بین میانگین دو گروه یا شرایط وجود دارد.</p>
</div>


```python
test , pval = stats.ttest_rel(df['stress_first'], df['stress_final'])
print('test :' , test)
print('P-value :' , pval)
```
